```python
#Dealing with Rows and column in pandas dataFrame
#we can perform bsic operation on rows/columns like selecting,deleting,adding and renaming.

#Example
#importing pandas
import pandas as pd

#Define a dictionary cotaining employee data
data = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Address':['Delhi', 'Kanpur', 'Allahabad', 'Kannauj'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']}

#convert the dictionary into DataFrame
df = pd.DataFrame(data)

#select two column
print(df[['Name','Address']])

      
```
