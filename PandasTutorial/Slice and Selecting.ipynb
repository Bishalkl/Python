```python
#Pandas Extracting rows using.loc[]

#importing the pandas module 
import pandas as pd

#Reading the csv
data = pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/nba.csv', index_col="Name")

#using loc[] method for get specific value 
first = data.loc["Raul Neto"]
second = data.loc["Jeff Withey"]

#using loc[] in mutiple parameter
third = data.loc[["R.J. Hunter","John Holland"]]

#checking datatype of rows
print(type(third))

#Extracting rows between two index labels
rows = data.loc["Shelvin Mack":"Jeff Withey"]
#data display
print(first,"\n\n\n",second)
print(third)
print(rows)
```

```python
#Exericse 

import pandas as pd

# Creating a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Age': [21, 22, 20, 23, 22],
    'Grade': [95, 89, 75, 92, 88],
    'Gender': ['F', 'M', 'M', 'M', 'F']
}

df = pd.DataFrame(data)
df.set_index('Name', inplace=True)

print("Original DataFrame:")
print(df)


#exerice one 

#first iloc
iloc_rows = df.iloc[2]
iloc_rows
print(iloc_rows)

#second loc
loc_rows = df.loc["Charlie"]
print(loc_rows)

#Exercise two 
e = df.loc[["Alice","David"],["Age","Gender"]]
print(e)

#Exercise three
f = df.iloc[:,[2]]
print(f)

#Exercise four
g = df.iloc[:,df[2]>85]
```
