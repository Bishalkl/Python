```python
# Calling on Series with n parameter()


#importing pandas module
import pandas as pd

#making data frame
data = pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/nba.csv')

# number of rows to return
n = 9

#creating series
series = data["Name"] #column "Name"

#returning top n rows
top = series.head(n = n)

#display
top
```

```python
#calling on series in with n parameter

#importing pandas module
import pandas as pd 

#making data frame
data= pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/nba.csv")

#numbers of rows  to return
n = 9

#creating series
series = data["Age"] # column of data

#calling tail() method
#storing in new variable 
data_bottom = series.tail(n = n)

#display 
data_bottom 
```
